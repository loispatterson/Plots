```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
import pandas as pd
import numpy as np
from PIL import Image
from collections import Counter
```

```python
import csv
 
# file object is created
file_ob = open(r"mum_words.txt")
 
# reader object is created
reader_ob = csv.reader(file_ob)
 
# contents of reader object is stored .
# data is stored in list of list format.
reader_contents = list(reader_ob)

# empty string is declare
text = ""
words = []
# iterating through list of rows
for row in reader_contents :
   # print(row)
    # concatenate the words
    words.append(row[0].lower())
```

```python
# Reads 'Youtube04-Eminem.csv' file
df = pd.read_csv(r"mum_words.txt")
 
word_count_dict = Counter(words)
#print(word_count_dict)
# The Image shape in which you wanna convert it to.

color_map2 = LinearSegmentedColormap.from_list('custom',
                                       [(0, '#F50E69'),
                                        (1,   '#560678')], N=20)


mask = np.array(Image.open(r'heartforlois.png'))
wordcloud = WordCloud(
            font_path='/Users/lois/Library/Fonts/arial 1.ttf',
            mask = mask,
            background_color ='white',
#            colormap='PuRd',
            colormap=color_map2,
            min_font_size = 10,
            max_font_size = 90,
            width = mask.shape[1],
            height = mask.shape[0]).generate_from_frequencies(word_count_dict)

#plot the WordCloud image                      
plt.figure(figsize = (10, 10), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)

plt.savefig('lois.png', dpi=600)
plt.show()
```
